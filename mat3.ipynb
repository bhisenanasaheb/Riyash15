```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
max_t = [ 10,20,30,40,50]
days = [ 1,2,3,4,5]
min_t = [15,30,31,25,37]
avr_t = [12,16,17,23,29]
        
```

```python
plt.plot(days,max_t,marker = "+",color="black", linestyle="--",markersize = "20" ,label= "max tem") # formating string
plt.plot(days,min_t,color="red",label="min temp")
plt.plot(days,avr_t,color="green", label =" avrg tem")
plt.xlabel("days")  #lebal and axies , grid
plt.ylabel("temprature")
plt.title("maha tempreture")
plt.legend(loc= "best" ,fontsize="small") # for suchi as label
plt.grid()
```
