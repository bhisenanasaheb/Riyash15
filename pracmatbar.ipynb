```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company = ["gogle","fb","ox","limb","apple"]
revenue = [ 85,56,45,63,58]
profit = [20,35,40,25,53]
m_price= [60,50,80,25,56]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue , width=0.4,label="REVENUE")
plt.bar(xpos+0.2,profit,width = 0.4,label="Profit")
plt.xlabel("revenue")
plt.ylabel("company name")
plt.title("company profit & revenue")
plt.legend()
```
